## Converting datatypes

```python
def not_float(value):
    try:
        f = float(value)
        return False
    except:
        return True
    
def non_float_values(df, col_name):
    return df.loc[df[col_name].apply(not_float), col_name]

def convert_str_to_int(df, col_name):
    try:
        df[col_name] = df[col_name].str.replace(r'[^0-9]', '').astype(int)
    except Exception as e:
        print('Could not convert', col_name)
```

```python
diary.info()
```

No null values

```python
non_float_values(diary, 'quantity')
```

```python
diary[diary.quantity.apply(not_float)].shape
```

We can filter these out using the handy ~ operator.

```python
diary = diary[~diary.quantity.apply(not_float)]
```

```python
diary.shape
```

Now we should be able to convert the column to float

```python
diary['quantity'] = diary.quantity.astype(float)
```

```python
diary[diary.calories.apply(not_float)]
```

```python
diary['calories'] = diary.calories.astype(float)
```

```python
diary['date'] = pd.to_datetime(diary.date)
```

```python
diary.info()
```

```python
daily.info()
```

```python
daily['date'] = pd.to_datetime(daily.date)
```

```python
non_float_values(daily, 'excercise_calories')
```

This occured because I applied the sum function to strings and it concatenated them. We should be able to fix with the eval function. we can also fix the spelling.

```python
daily['exercise_calories'] = daily.excercise_calories.apply(eval)
daily.drop('excercise_calories', axis=1, inplace=True)
```

We can also drop the columns that had no target

```python
daily.drop(['Starch', 'Sugars', 'Monounsaturated', 'Polyunsaturated', 'Saturated', 'Trans-Fats',
           'Cholesterol'], axis=1, inplace=True)
```

Now we should be able to convert the remaining columns to integer. For example 123% -> 123

```python
for col in daily.columns[3:43]:
    convert_str_to_int(daily, col)
```
